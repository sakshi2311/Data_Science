```
#creating a new corpus variable called corpus
corpus = df.copy()
```

```
#convert the text to lowercase. 
for i in range(len(corpus)):
  corpus['text'][i] = corpus['text'][i].lower()
corpus.head()
```

```
import re
for i in range(len(corpus)):
  corpus["text"][i] = re.sub(r'[^\w\s]', '', corpus["text"][i])
```

```
from nltk.tokenize import word_tokenize

for i in range(len(corpus)):
  corpus['text'][i] = word_tokenize(corpus['text'][i])
  corpus["text"][i] = [w for w in corpus["text"][i] if not w.lower() in stop_words_maam]

corpus.head()  
```

```
#Stemming

from nltk.stem import PorterStemmer
from nltk.stem.snowball import SnowballStemmer

ps = PorterStemmer()
sb = SnowballStemmer("english")

for i in range(len(df)):
    stemmed = []
    for ele in corpus['text'][i]:
        ele = ps.stem(ele)
        stemmed.append(ele)
    corpus['text'][i] = stemmed
```
